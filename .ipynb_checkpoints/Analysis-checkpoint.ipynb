```python
import matplotlib.pyplot as plt
for country in cosmetics['geo'].unique(): 
    subset=cosmetics[cosmetics['geo']==country]
    plt.plot(subset['TIME_PERIOD'], subset['OBS_VALUE'], label=country) 
plt.legend()
plt.title("CosmeticsnSales Index (2021=100)")
plt.xlabel("Time")
plt.ylabel("Index")
plt.show() 
```

```python
cosmetics=cosmetics.copy()
cosmetics['TIME_PERIOD']=pd.to_datetime(cosmetics['TIME_PERIOD'])
plt.figure(figsize=(10,5))
for country in cosmetics['geo'].unique():
    subset=cosmetics[cosmetics['geo']==country].sort_values('TIME_PERIOD')
    plt.plot(subset['TIME_PERIOD'], subset['OBS_VALUE'], label=country)
plt.legend()
plt.title("Cosmetics Sale Index (2021=100)")
plt.xlabel("Time")
plt.ylabel("Index")
plt.show()
```

```python
plt.figure(figsize=(10,5))
for country in cosmetics['geo'].unique():
    subset=cosmetics[cosmetics['geo']==country].sort_values('TIME_PERIOD') 
    rolling=subset['OBS_VALUE']. rolling(window=12).mean()
    plt.plot(subset['TIME_PERIOD'], rolling, label=country)
plt.legend()
plt.title("Cosmetics Sales Index - 12 Month Rolling Average")
plt.xlabel("Time")
plt.ylabel("Index")
plt.show() 
```

```python
import pandas as pd

#Merge cosmetics and unemployment data on country and time
cosmetics_avg = cosmetics.groupby(['geo', 'TIME_PERIOD'])['OBS_VALUE'].mean().reset_index()
cosmetics_avg.columns=['geo','TIME_PERIOD','cosmetics_index']

unemp_clean=unemp[['geo', 'TIME_PERIOD', 'OBS_VALUE']].copy()
unemp_clean.columns=['geo', 'TIME_PERIOD', 'unemployment_rate']

merged=pd.merge(cosmetics_avg, unemp_clean, on=['geo', 'TIME_PERIOD']) 

#Calculate correlation for each country 
for country in ['ES', 'FR', 'IT']:
    subset=merged[merged['geo']==country]
    corr=subset['cosmetics_index'].corr(subset['unemployment_rate'])
    print(f"{country}:correlation={corr:.2f}") 
```

```python
unemp=pd.read_csv("data/unemployment.csv")
unemp['TIME_PERIOD']=pd.to_datetime(unemp['TIME_PERIOD'])
```

```python
import matplotlib.pyplot as plt 

fig,axes=plt.subplots(1, 3, figsize=(14,4))
countries = {'ES':'Spain', 'FR':'France', 'IT': 'Italy'}

for ax, (code, name) in zip(axes, countries.items()):
    subset=merged[merged['geo']==code]
    ax.scatter(subset['unemployment_rate'], subset['cosmetics_index'], alpha=0.4, s=10)
    ax.set_title(name)
    ax.set_xlabel("Unemployment Rate (%)")
    ax.set_ylabel("Cosmetics Sales Index")

plt.suptitle("Cosmetics Sales vs Unemployment Rate", fontsize=13)
plt.tight_layout()
plt.savefig("images/scatter_plot.png", dpi=150)
plt.show()
```
